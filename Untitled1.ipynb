Model Evaluation

Regression Evaluation Metrics:

These model evaluation techniques are used to find out the accuracy of models built in the Regression type of machine learning models. We have three types of evaluation methods.

R-square_score
RMSE  root mean squared error

1. R-squared _score -



### R-squared (Coefficient of Determination)

The R-squared score is calculated using the formula:

\[
R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}
\]

Where:
- \( y_i \) = Actual value  
- \( \hat{y}_i \) = Predicted value  
- \( \bar{y} \) = Mean of actual values  
- \( n \) = Total number of samples


```python
print(metrics.r2_score(p1,y_train))
print(metrics.r2_score(p2,y_train))
print(metrics.r2_score(p3,y_train))
print(metrics.r2_score(p4,y_train))
print(metrics.r2_score(p5,y_train))
```

```python
p1 = lin_reg.predict(x_test)
p2 = D_tree.predict(x_test)
p3 = rand.predict(x_test)
p4 = svr.predict(x_test)
p5 = XGB.predict(x_test)
```

```python
print(metrics.r2_score(p1,y_test))
print(metrics.r2_score(p2,y_test))
print(metrics.r2_score(p3,y_test))
print(metrics.r2_score(p4,y_test))
print(metrics.r2_score(p5,y_test))
```

MSE Root Mean Square Error

### RMSE  Root Mean Square Error

RMSE measures the standard deviation of the prediction errors (residuals). It is calculated as:

\[
\text{RMSE} = \sqrt{ \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 }
\]

Where:
- \( y_i \) = Actual value  
- \( \hat{y}_i \) = Predicted value  
- \( n \) = Total number of observations


```python
rmse_lin = np.sqrt(mean_squared_error(y_test, p1))
rmse_tree = np.sqrt(mean_squared_error(y_test, p2))
rmse_rand = np.sqrt(mean_squared_error(y_test, p3))
rmse_svr = np.sqrt(mean_squared_error(y_test, p4))
rmse_xgb = np.sqrt(mean_squared_error(y_test, p5))
```

```python
print("Linear Regression RMSE:", rmse_lin)
print("Decision Tree RMSE:", rmse_tree)
print("Random Forest RMSE:", rmse_rand)
print("SVR RMSE:", rmse_svr)
print("XGBoost RMSE:", rmse_xgb)
```

```python
print(np.sqrt(rmse_lin))
print(np.sqrt(rmse_tree))
print(np.sqrt(rmse_rand))
print(np.sqrt(rmse_svr))
print(np.sqrt(rmse_xgb))
```
